## ðŸ’¾ The data: sleep_health_data.csv

SleepInc has provided you with an anonymized dataset of sleep and lifestyle metrics for 374 individuals. This dataset contains average values for each person calculated over the past six months. The data is saved as `sleep_health_data.csv`.

The dataset includes 13 columns covering sleep duration, quality, disorders, exercise, stress, diet, demographics, and other factors related to sleep health. 

| Column | Description |
|---------|----------------------------------------|  
| `Person ID` | An identifier for each individual. |
| `Gender` | The gender of the person (Male/Female). |  
| `Age` | The age of the person in years. |
| `Occupation` | The occupation or profession of the person. |
| `Sleep Duration (hours)` | The average number of hours the person sleeps per day. |
| `Quality of Sleep (scale: 1-10)` | A subjective rating of the quality of sleep, ranging from 1 to 10. |
| `Physical Activity Level (minutes/day)` | The average number of minutes the person engages in physical activity daily. |  
| `Stress Level (scale: 1-10)` | A subjective rating of the stress level experienced by the person, ranging from 1 to 10. |
| `BMI Category` | The BMI category of the person (e.g., Underweight, Normal, Overweight). |
| `Blood Pressure (systolic/diastolic)` | The average blood pressure measurement of the person, indicated as systolic pressure over diastolic pressure. |
| `Heart Rate (bpm)` | The average resting heart rate of the person in beats per minute. |
| `Daily Steps` | The average number of steps the person takes per day. |
| `Sleep Disorder` | The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea). |

```python
# Start coding here
# Use as many cells as you need
import pandas as pd
```

```python
df= pd.read_csv("sleep_health_data.csv")
df.head()
```

### Find occupation with the lowest sleep duration.

```python
#Group the occupations, then calculate the mean sleep duration for each occupation
sleep_duration=df.groupby("Occupation")["Sleep Duration"].mean()
sleep_duration
```

```python
#the occupation with the lowest sleep duration.
lowest_sleep_occ= sleep_duration.sort_values().index[0]
lowest_sleep_occ
```

### Find occupation with the lowest sleep quality.

```python
#Group the occupations, then calculate the mean sleep quality for each occupation
sleep_quality= df.groupby("Occupation")["Quality of Sleep"].mean()
sleep_quality
```

```python
#occupation with the lowest sleep quality.
lowest_sleep_quality_occ= sleep_quality.sort_values().index[0]
lowest_sleep_quality_occ
```

```python
# Check if the same occupation has the lowest average sleep duration and quality.
if lowest_sleep_occ == lowest_sleep_quality_occ:
    same_occ= True
else:
    same_occ=False
        
same_occ
```

### Find what ratio of app users in each BMI category have been diagnosed with Insomnia.

```python
df["BMI Category"].unique()
```

```python
#normal insomnia
normal_insomnia= df[(df["BMI Category"]== "Normal") & (df["Sleep Disorder"]== "Insomnia")]

#obese insomnia
obese_insomnia= df[(df["BMI Category"]== "Obese") & (df["Sleep Disorder"]== "Insomnia")]

#overweight insomnia
overweight_insomnia= df[(df["BMI Category"]== "Overweight") & (df["Sleep Disorder"]== "Insomnia")]
```

```python
bmi_insomnia_ratios={
    "Normal": float,
    "Overweight": float,
    "Obese": float
}
```

```python
#normal ratio
total_normal =len(df[df["BMI Category"]== "Normal"])

normal_ratio=round(len(normal_insomnia)/total_normal, 2)

bmi_insomnia_ratios["Normal"]=normal_ratio
```

```python
#obese ratio
total_obese =len(df[df["BMI Category"]== "Obese"])

obese_ratio=round(len(obese_insomnia)/total_obese, 2)

bmi_insomnia_ratios["Obese"]=obese_ratio
```

```python
#overweight ratio
total_overweight =len(df[df["BMI Category"]== "Overweight"])

overweight_ratio=round(len(overweight_insomnia)/total_overweight, 2)

bmi_insomnia_ratios["Overweight"]=overweight_ratio
```
